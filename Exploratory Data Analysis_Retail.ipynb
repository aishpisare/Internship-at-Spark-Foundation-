```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("SampleSuperstore.csv")
data.head()
```

```python
data.isna().sum()
```

```python
data.info()
```

# EDA

```python
plt.figure(figsize=(35,15))
ed=sns.countplot(x="State", data=data)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('State', fontsize=30) 
plt.ylabel('Count', fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(35,15))
state = data.groupby(['State'])['Sales'].mean().reset_index()
ed=sns.barplot(x = 'State', y='Sales', data=state)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('State', fontsize=30) 
plt.ylabel('Mean Sales', fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(25,10))
ed=sns.countplot(x="Category", data=data)
ed.set_xticklabels(ed.get_xticklabels(),rotation=90)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Category', fontsize=20) 
plt.ylabel('Count', fontsize=20)
plt.show()
```

```python

plt.figure(figsize=(25,10))
category = data.groupby(['Category'])['Sales'].mean().reset_index()
ed=sns.barplot(x = 'Category', y='Sales', data=category)
ed.set_xticklabels(ed.get_xticklabels())
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Category', fontsize=20) 
plt.ylabel('Mean Sales', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(25,10))
category1 = data.groupby(['Category'])['Discount'].mean().reset_index()
ed=sns.barplot(x = 'Category', y='Discount', data=category1)
ed.set_xticklabels(ed.get_xticklabels())
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Category', fontsize=20) 
plt.ylabel('Mean Discount', fontsize=20)
plt.show()
```

```python

plt.figure(figsize=(25,10))
category2 = data.groupby(['Category'])['Profit'].mean().reset_index()
ed=sns.barplot(x = 'Category', y='Profit', data=category2)
ed.set_xticklabels(ed.get_xticklabels())
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Category', fontsize=20) 
plt.ylabel('Mean Profit', fontsize=20)
plt.show()
```
